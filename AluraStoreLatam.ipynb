### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingresos_de_las_tienda = {
     "tienda_1":tienda['Precio'].sum(),
     "tienda_2":tienda2['Precio'].sum(),
     "tienda_3":tienda3['Precio'].sum(),
     "tienda_4":tienda4['Precio'].sum()
 }

```

```python
df_ingresos = pd.DataFrame(list(ingresos_de_las_tienda.items()), columns=['Tienda','Ingreso total'])
```

```python
df_ingresos
```

# 2. Ventas por categoría

```python
tienda_ventas = {
    "tienda_1":tienda['Categoría del Producto'].value_counts(),
    "tienda_2":tienda2['Categoría del Producto'].value_counts(),
    "tienda_3":tienda3['Categoría del Producto'].value_counts(),
    "tienda_4":tienda4['Categoría del Producto'].value_counts()
}
```

```python
tienda_ventas
```

```python
df_venta_categoria = pd.DataFrame(tienda_ventas)
```

```python
df_venta_categoria
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion = {
    "tienda_1":tienda['Calificación'].mean(),
    "tienda_2":tienda2['Calificación'].mean(),
    "tienda_3":tienda3['Calificación'].mean(),
    "tienda_4":tienda4['Calificación'].mean()
}
```

```python
df_promedio_calificacion = pd.DataFrame(list(promedio_calificacion.items()), columns=['Tienda','Calificación promedio'])
```

```python
df_promedio_calificacion
```

# 4. Productos más y menos vendidos

```python
mas_vendidos = {
    "tienda_1":tienda['Producto'].value_counts(),
    "tienda_2":tienda2['Producto'].value_counts(),
    "tienda_3":tienda3['Producto'].value_counts(),
    "tienda_4":tienda4['Producto'].value_counts()
}
```

```python
df_mas_vendidos = pd.DataFrame(mas_vendidos).fillna(0).astype(int)
display(df_mas_vendidos)
```

# 5. Envío promedio por tienda

```python
promedio_envio = {
    "tienda_1":tienda['Costo de envío'].mean(),
    "tienda_2":tienda2['Costo de envío'].mean(),
    "tienda_3":tienda3['Costo de envío'].mean(),
    "tienda_4":tienda4['Costo de envío'].mean()
}

df_promedio_envio = pd.DataFrame(list(promedio_envio.items()), columns=['Tienda','Costo promedio de envío'])
```

```python
display(df_promedio_envio)
```

#**Generando graficos**

```python
#Gráfico de barras — Ingresos por tienda
df_ingresos.plot.bar(x='Tienda', y='Ingreso total', title='Ingresos por tienda')
```

```python
#Gráfico circular — Distribución de categorías
df_venta_categoria.plot.pie(subplots=True, figsize=(12, 14))
```

```python
#Gráfico de líneas — Promedio del costo de envío
df_promedio_envio.plot.line(x='Tienda', y='Costo promedio de envío', title='Promedio del costo de envío')
```

#**Informe Final**
##Introducción

El objetivo de este análisis fue determinar cuál de las cuatro tiendas — tienda_1, tienda_2, tienda_3 y tienda_4 — es la mejor opción para que el Sr. Juan venda sus productos.
Se estudiaron los ingresos totales, la distribución de categorías, los productos más vendidos y el promedio del costo de envío.
Los resultados se visualizaron mediante gráficos para identificar patrones y oportunidades.

#**Desarrollo del Análisis**
##**1. Ingresos totales por tienda**

El gráfico de barras muestra que todas las tiendas tienen ingresos similares, aunque:

Tienda 1 registra el mayor ingreso total (≈ 1.15 × 10⁹).

Tienda 4 presenta el menor ingreso (≈ 1.03 × 10⁹).

Esto sugiere que Tienda 1 tiene el mejor desempeño económico general.

##**2. Distribución de categorías de productos**

Los gráficos circulares indican que:

Las categorías con mayor participación en todas las tiendas son Electrónica, Artículos para el hogar y Electrodomésticos.

Las menos frecuentes son Libros, Instrumentos musicales y Muebles.

Esto revela que las tiendas se concentran en productos tecnológicos y de uso doméstico, lo que coincide con las preferencias de los compradores.

##**3. Promedio del costo de envío**

El gráfico de líneas muestra una tendencia descendente:

Tienda 1 tiene el costo de envío más alto (≈ 26 000).

Tienda 4 tiene el más bajo (≈ 23 000).

Aunque las diferencias no son extremas, los envíos más baratos en la Tienda 4 podrían atraer más clientes sensibles al precio.

#**Conclusión y Recomendación**

Al combinar los indicadores:

Tienda 1 destaca por sus mayores ingresos, su variedad de productos y su fuerte participación en categorías populares.

Tienda 4, aunque tiene el menor costo de envío, genera menos ingresos y parece tener un volumen de ventas más bajo.

Las tiendas 2 y 3 se ubican en posiciones intermedias en todos los aspectos.

Recomendación final:

El Sr. Juan debería vender sus productos en Tienda 1, ya que combina altos ingresos, una base sólida de clientes y una buena diversificación de productos.
Si el factor principal fuera el costo logístico, Tienda 4 podría considerarse como alternativa, pero con menor potencial de ventas.
